### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
loja2.head()
loja3.head()
loja4.head()
```

#1. Análise do faturamento


```python
def calcular_faturamento(loja_df, nome_loja):
    faturamento = loja_df['Preço'].sum() - loja_df['Frete'].sum()
    print(f"O faturamento da {nome_loja} foi de R${faturamento:,.2f}".replace(',', 'v').replace('.', ',').replace('v', '.'))
    return faturamento

# ---------- GRÁFICO 1: Faturamento por loja ----------
faturamentos = {
    "Loja 1": calcular_faturamento(loja, "loja 1"),
    "Loja 2": calcular_faturamento(loja2, "loja 2"),
    "Loja 3": calcular_faturamento(loja3, "loja 3"),
    "Loja 4": calcular_faturamento(loja4, "loja 4"),
}

plt.figure(figsize=(8, 5))
plt.bar(faturamentos.keys(), faturamentos.values(), color='teal')
plt.title("Faturamento por Loja")
plt.ylabel("R$")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
def mostrar_total_por_categoria(loja_df, nome_loja):
    total_por_categoria = loja_df.groupby('Categoria do Produto')['Preço'].count()
    print(f'''
{nome_loja}
{total_por_categoria.sort_values(ascending=False)}
''')
    return total_por_categoria


# Gráfico Loja 1
categorias_loja1 = mostrar_total_por_categoria(loja, "Loja 1")
plt.figure(figsize=(6, 6))
plt.pie(categorias_loja1, labels=categorias_loja1.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribuição por Categoria (Loja 1)")
plt.axis('equal')
plt.tight_layout()
plt.show()

# Gráfico Loja 2
categorias_loja2 = mostrar_total_por_categoria(loja2, "Loja 2")
plt.figure(figsize=(6, 6))
plt.pie(categorias_loja2, labels=categorias_loja2.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribuição por Categoria (Loja 2)")
plt.axis('equal')
plt.tight_layout()
plt.show()

# Gráfico Loja 3
categorias_loja3 = mostrar_total_por_categoria(loja3, "Loja 3")
plt.figure(figsize=(6, 6))
plt.pie(categorias_loja3, labels=categorias_loja3.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribuição por Categoria (Loja 3)")
plt.axis('equal')
plt.tight_layout()
plt.show()

# Gráfico Loja 4
categorias_loja4 = mostrar_total_por_categoria(loja4, "Loja 4")
plt.figure(figsize=(6, 6))
plt.pie(categorias_loja4, labels=categorias_loja4.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribuição por Categoria (Loja 4)")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
print(f"A loja 1 está com uma avaliação média de: {round(loja['Avaliação da compra'].mean(), 2)}")
print(f"A loja 2 está com uma avaliação média de: {round(loja2['Avaliação da compra'].mean(), 2)}")
print(f"A loja 3 está com uma avaliação média de: {round(loja3['Avaliação da compra'].mean(), 2)}")
print(f"A loja 4 está com uma avaliação média de: {round(loja4['Avaliação da compra'].mean(), 2)}")

avaliacoes = {
    "Loja 1": loja['Avaliação da compra'].mean(),
    "Loja 2": loja2['Avaliação da compra'].mean(),
    "Loja 3": loja3['Avaliação da compra'].mean(),
    "Loja 4": loja4['Avaliação da compra'].mean(),
}

plt.figure(figsize=(8, 4))
plt.plot(list(avaliacoes.keys()), list(avaliacoes.values()), marker='o', linestyle='-', color='purple')
plt.title("Avaliação Média por Loja")
plt.ylabel("Nota média")
plt.ylim(0, 5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
def mostrar_ranking_produtos(loja_df, nome_loja):
    ranking = loja_df.groupby('Produto')['Produto'].count().sort_values(ascending=False)
    print(f"\nProdutos mais vendidos na {nome_loja}:\n")
    print(ranking)
    print(f"\nMais vendido: {ranking.idxmax()} ({ranking.max()} vendas)")
    print(f"Menos vendido: {ranking.idxmin()} ({ranking.min()} vendas)")


mostrar_ranking_produtos(loja, "Loja 1")
mostrar_ranking_produtos(loja2, "Loja 2")
mostrar_ranking_produtos(loja3, "Loja 3")
mostrar_ranking_produtos(loja4, "Loja 4")
```

# 5. Frete Médio por Loja


```python
print(f"A loja 1 está com um frete médio de: R${round(loja['Frete'].mean(), 2)}")
print(f"A loja 2 está com um frete médio de: R${round(loja2['Frete'].mean(), 2)}")
print(f"A loja 3 está com um frete médio de: R${round(loja3['Frete'].mean(), 2)}")
print(f"A loja 4 está com um frete médio de: R${round(loja4['Frete'].mean(), 2)}")



avaliacoes = {
    "Loja 1": round(loja['Frete'].mean(), 2),
    "Loja 2": round(loja2['Frete'].mean(), 2),
    "Loja 3": round(loja3['Frete'].mean(), 2),
    "Loja 4": round(loja4['Frete'].mean(), 2),
}

plt.figure(figsize=(8, 4))
plt.plot(list(avaliacoes.keys()), list(avaliacoes.values()), marker='o', linestyle='-', color='purple')
plt.title("Frete Médio por Loja")
plt.ylabel("Nota média")
plt.ylim(0, 50)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
